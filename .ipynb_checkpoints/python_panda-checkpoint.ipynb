### Pegar 1 coluna (e os pd.Series)

```python
produtos = vendas_sp[['Produto', 'ID Loja']]
display(produtos)
```

### .loc, um método muito importante
    - Pegar 1 linha
    - Pegar linhas de acordo com alguma condição
    - Pegar linhas e colunas específicas
    - Pegar 1 valor específico

```python
# pegar uma linha
display(vendas_sp.loc[1:5])

# pegar linhas que corresponde a uma condição
vendas_norteshopping_sp = vendas_sp.loc[vendas_sp['ID Loja'] == 'Norte Shopping']
display(vendas_norteshopping_sp)

#pegar várias linhas e colunas usando loc
#loc[linhas, colunas]
vendas_norteshopping_sp = vendas_sp.loc[vendas_sp['ID Loja'] == 'Norte Shopping', ["ID Loja", "Produto", "Quantidade"]]
display(vendas_norteshopping_sp)

# pegar 1 valor específico
print(vendas_sp.loc[1, 'Produto'])
```

### Adicionar 1 coluna

```python
# Adicionando coluna apartir de uma coluna que já existe
vendas_sp["Comissão"] = vendas_sp["Valor Final"] * 0.05
display(vendas_sp)
```

### Adicionar 1 linha
    - Linhas de um complemento da base de dados

```python
vendas_dez_sp = pd.read_excel("Vendas - Dez.xlsx")

vendas_sp = vendas_sp.append(vendas_dez_sp)
display(vendas_sp)
```

### Excluir linhas e colunas

```python
vendas_sp = vendas_sp.drop(0, axis=0)
display(vendas_sp)
```

### Valores Vazios
    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e último valor)

```python
#deletar vendas e colunas completamente vazias
#vendas_sp = vendas_sp.dropna(how='all', axis=1)
#display(vendas_sp)

#delerar linhas que possuem pelo menos 1 valor vazio
#vendas_sp = vendas_sp.dropna()
#display(vendas_sp)

#preencher valores vazios 
#preencher com média da coluna
vendas_sp["Comissão"] = vendas_sp["Comissão"].fillna(vendas_sp["Comissão"].mean())
display(vendas_sp)


#preencher com último valor
vendas_sp = vendas_sp.ffill()
display(vendas_sp)
```

### Calcular Indicadores
    - Groupby
    - Value Counts

```python
# value counts
#transicoes_loja = vendas_sp['ID Loja'].value_counts()
#display(transicoes_loja)

# group by
faturamento_produto = vendas_sp[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

### Mesclar 2 dataframes (Procurar informações de um dataframe em outro)

```python
gerentes_sp = pd.read_excel("Gerentes.xlsx")
#display(gerentes_sp)

vendas_sp = vendas_sp.merge(gerentes_sp)
display(vendas_sp)

```
